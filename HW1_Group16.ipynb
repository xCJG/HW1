```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

def loadDataset(url):
    df = pd.read_csv(url, header=None)
    array = df.to_numpy()
    X = array[:, :-1]
    y = array[:, -1]
    return X, y

def evaluate_knn(X_train, y_train, X_test, y_test, k_values, repetitions):
    results = {k: [] for k in k_values}

    for k in k_values:
        for _ in range(repetitions):
            # Initialize KNeighborsClassifier
            model = KNeighborsClassifier(n_neighbors=k)
            model.fit(X_train, y_train)
            
            # Predict on test data
            predictions = model.predict(X_test)
            
            # Calculate accuracy
            accuracy = accuracy_score(y_test, predictions)
            results[k].append(accuracy)
    
    # Calculate average accuracy for each k
    avg_results = {k: np.mean(v) for k, v in results.items()}
    return avg_results

def plot_accuracy(avg_accuracies):
    ks = list(avg_accuracies.keys())
    accuracies = list(avg_accuracies.values())
    
    plt.figure(figsize=(10, 6))
    plt.plot(ks, accuracies, marker='o', linestyle='-', color='b')
    plt.title('Average Accuracy vs. Number of Neighbors (k)')
    plt.xlabel('Number of Neighbors (k)')
    plt.ylabel('Average Accuracy')
    plt.grid(True)
    plt.xticks(ks)
    plt.show()

def main():
    url = 'https://raw.githubusercontent.com/ruiwu1990/CSCI_4120/master/KNN/iris.data'
    
    # Load dataset
    X, y = loadDataset(url)
    
    # Split data into training and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
    
    # Define k values and repetitions
    k_values = range(1, 21)
    repetitions = 5
    
    # Evaluate kNN
    avg_accuracies = evaluate_knn(X_train, y_train, X_test, y_test, k_values, repetitions)
    
    # Print results
    for k, accuracy in avg_accuracies.items():
        print(f'k={k}, Average Accuracy={accuracy:.2f}%')
    
    # Plot the results
    plot_accuracy(avg_accuracies)

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

def loadDataset(url):
    df = pd.read_csv(url, header=None)
    array = df.to_numpy()
    X = array[:, :-1]
    y = array[:, -1]
    return X, y

def evaluate_knn(X_train, y_train, X_test, y_test, k_values, repetitions):
    results = {k: [] for k in k_values}

    for k in k_values:
        for _ in range(repetitions):
            # Initialize KNeighborsClassifier
            model = KNeighborsClassifier(n_neighbors=k)
            model.fit(X_train, y_train)
            
            # Predict on test data
            predictions = model.predict(X_test)
            
            # Calculate accuracy
            accuracy = accuracy_score(y_test, predictions)
            results[k].append(accuracy)
    
    # Calculate average accuracy for each k
    avg_results = {k: np.mean(v) for k, v in results.items()}
    return avg_results

def plot_accuracy(avg_accuracies):
    ks = list(avg_accuracies.keys())
    accuracies = list(avg_accuracies.values())
    
    plt.figure(figsize=(10, 6))
    plt.plot(ks, accuracies, marker='o', linestyle='-', color='b')
    plt.title('Average Accuracy vs. Number of Neighbors (k)')
    plt.xlabel('Number of Neighbors (k)')
    plt.ylabel('Average Accuracy')
    plt.grid(True)
    
    # Customize y-axis ticks to go up by 0.5
    min_accuracy = min(accuracies)
    max_accuracy = max(accuracies)
    plt.yticks(np.arange(np.floor(min_accuracy), np.ceil(max_accuracy) + 0.5, 0.5))
    
    plt.xticks(ks)
    plt.show()

def main():
    url = 'https://raw.githubusercontent.com/ruiwu1990/CSCI_4120/master/KNN/iris.data'
    
    # Load dataset
    X, y = loadDataset(url)
    
    # Split data into training and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
    
    # Define k values and repetitions
    k_values = range(1, 21)
    repetitions = 5
    
    # Evaluate kNN
    avg_accuracies = evaluate_knn(X_train, y_train, X_test, y_test, k_values, repetitions)
    
    # Print results
    for k, accuracy in avg_accuracies.items():
        print(f'k={k}, Average Accuracy={accuracy:.2f}%')
    
    # Plot the results
    plot_accuracy(avg_accuracies)

if __name__ == "__main__":
    main()
```
